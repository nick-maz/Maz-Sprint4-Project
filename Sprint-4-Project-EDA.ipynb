```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('vehicles_us.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Changing the data type in the is_4wd column to str and replacing 1.0 with 'yes' and Nan with 'no'
conversion_dict = {1.0: 'yes', None: 'no'}
df['is_4wd'] = df['is_4wd'].map(conversion_dict)
display(df.head())
```

```python
# Removing missing values in the model_year column 
df.dropna(subset=['model_year'], inplace=True)
display(df.head())
```

```python
# Changing the data type of the model_year columnm to int
df['model_year'] = df['model_year'].astype(int)
display(df.info())
display(df.head())
```

```python
# Filling missing values in the odometer column with 0
df['odometer'] = df['odometer'].fillna(0)
display(df['odometer'].head())
```

```python
# Changing the data type of the odometer column to int
df['odometer'] = df['odometer'].astype(int)
display(df.head())
```

```python
# Filling missing values in the paint_color column with "not found"
df['paint_color'] = df['paint_color'].fillna('not found')
display(df.head())
```

```python
# Checking for missing values in the cylinders column
df['cylinders'].isna()
```

```python
# Filling missing values with 0 in the cylinders column
df['cylinders'] = df['cylinders'].fillna(0)
display(df.head())
```

```python
#Changing the data type of the cylinders column to int
df['cylinders'] = df['cylinders'].astype(int)
display(df.head())
```
